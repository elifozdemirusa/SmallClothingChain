```python
# Exploratory Data Analysis
# Small Clothing Store Chain
# Author: Elif Ozdemir

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample transaction data (placeholder for EDA)

transaction_data = {
    "transaction_id": [1, 2, 3, 4, 5],
    "customer_id": [101, 102, 101, 103, 104],
    "employee_id": [201, 202, 201, 203, 202],
    "total": [120.50, 75.00, 200.00, 50.25, 180.75],
    "transaction_date": [
        "2024-01-15",
        "2024-01-20",
        "2024-02-05",
        "2024-02-18",
        "2024-03-02"
    ]
}

transaction_df = pd.DataFrame(transaction_data)

# Convert date column to datetime
transaction_df["transaction_date"] = pd.to_datetime(transaction_df["transaction_date"])

transaction_df
```

```python
# Monthly sales analysis

transaction_df['month'] = transaction_df['transaction_date'].dt.to_period('M').astype(str)

monthly_sales = transaction_df.groupby('month')['total'].sum()

plt.figure(figsize=(8, 4))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Total Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.tight_layout()
plt.show()

# Insight:
# Sales show month-to-month variation, suggesting potential seasonality
# or changes in customer purchasing be
```

```python
# Employee sales performance

employee_sales = (
    transaction_df
    .groupby('employee_id')['total']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8, 4))
employee_sales.plot(kind='bar')
plt.title("Total Sales by Employee")
plt.xlabel("Employee ID")
plt.ylabel("Total Sales ($)")
plt.tight_layout()
plt.show()

# Insight:
# Sales contribution varies significantly across employees,
# indicating differences in performance that could inform training
# or incentive programs.
```
